### Задание 8 / Task 8
В цилиндрический сосуд налили $2100 \ см^3$ воды. Уровень жидкости оказался равным 20 см.

В воду полностью погрузили деталь. При этом уровень жидкости в сосуде поднялся на 5 см.

Чему равен объем детали? Ответ выразите в $см^3$.

---

$2100 \ cm^3$ of water were poured into a cylindrical vessel. The liquid level was equal to 20 cm.

The part component was completely immersed in the water. At the same time, the liquid level in the vessel rose by 5 cm.

What is the volume of a part? The answer should be expressed in $cm^3$.

```python
f = plt.figure(figsize=(9,9))
ax = f.add_subplot(111, projection='3d')

def create_cylinder(center_x,center_y,radius,height):
    z = numpy.linspace(0, height, 50)
    theta = numpy.linspace(0, 2*numpy.pi, 50)
    theta_grid, z_grid = numpy.meshgrid(theta, z)
    x_grid = radius * numpy.cos(theta_grid) + center_x
    y_grid = radius * numpy.sin(theta_grid) + center_y
    return x_grid,y_grid,z_grid

Xc,Yc,Zc = create_cylinder(0,0,0.1,0.2)
ax.plot_surface(Xc, Yc, Zc, alpha=0.7, color='darkblue')

Xc2,Yc2,Zc2 = create_cylinder(0,0,0.1,0.25)
ax.plot_surface(Xc2, Yc2, Zc2, alpha=0.3, color='darkblue')

Xc3,Yc3,Zc3 = create_cylinder(0,0,0.1,0.3)
ax.plot_surface(Xc3, Yc3, Zc3, alpha=0.1, color='steelblue')

plt.axis('off');
```

$V_1 = \pi ∗ r^2 ∗ h_1 \implies  \pi ∗ r^2 = V_1 / h_1$

$V_2 = \pi ∗ r^2 ∗ h_2 = V_1 / h_1 * h_2$

$V_2 = 2100 / 20 * 5 = 525$

### Ответ / Answer: 525

```python
answer_dictionary[8]=str(int(2100/20*5))
```
